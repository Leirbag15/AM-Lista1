# 4 - Correlação
**Direto ao ponto**: correlação é o coeficiente de interdependência entre duas ou mais variáveis.

Em muitos casos, quando analisamos algumas variáveis, essa interdependência é muito clara. Exemplo:
- Massa corporal e altura,
- Consumo de combustível e distância percorrida,
- Número de anúncios e produtos vendidos,
- Tempo de estudo e desempenho na avaliação.

Entretanto, algumas vezes essa correlação pode não ser muito clara.

## Correlações positivas

```python
import numpy as np

np.random.seed(1)

# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Positive Correlation with some noise
y = x + np.random.normal(0, 10, 1000)

np.corrcoef(x, y)
```

```python
# import matplotlib
import matplotlib.pyplot as plt

plt.scatter(x, y)
```

## Correlações negativas

```python
# 1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)

# Negative Correlation with some noise
y = 100 - x + np.random.normal(0, 5, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

## Correlação fraca ou sem correlação

```python
x = np.random.randint(0, 50, 1000)
y = np.random.randint(0, 50, 1000)

np.corrcoef(x, y)
```

```python
plt.scatter(x, y)
```

```python
#1
x = np.random.randint(0, 50, 1000)
y = x + np.random.randint(0,15,1000)
print(np.corrcoef(x, y))
plt.scatter(x, y)
#Correlações positivas são encontradas em situações que o aumento no valor de uma variável indica o aumento no valor de outra
```

```python
#2
x = np.random.randint(0,50,1000)
y = 100 - x + np.random.randint(0,10,1000)
print(np.corrcoef(x, y))
plt.scatter(x, y)
#correlações negativas são situações em que o aumento no valor de uma variável indicam uma diminuição no valor de outro
```

```python
#3
x = np.random.randint(0,100,1000)
y = np.random.randint(0,100,1000)
print(np.corrcoef(x, y))
plt.scatter(x, y)
#Pode acontecer de duas variáveis serem completamente independentes entre si, e nesse caso não há correlação
```

## Correlação é diferente de causalidade!
